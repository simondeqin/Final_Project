### Train, Validation, Test Set 

```python
def create_train_test_set(scaled_set_df):
    
    features = scaled_set_df.drop(columns=['BRK_Close'], axis=1)
    target = scaled_set_df['BRK_Close']
    

    data_len = scaled_set_df.shape[0]
    print('Historical Stock Data length is:', str(data_len))

    #create a chronological split for train and testing
    train_split = int(data_len * 0.88)
    print('Training Set length:', str(train_split))

    val_split = train_split + int(data_len * 0.1)
    print('Validation Set length:', str(int(data_len * 0.1)))

    print('Test Set length:', str(int(data_len * 0.02)))

    # Splitting features and target into train, validation and test samples
    ## ATTENTION! The Y should be the next day, not the same day as X
    X_train, X_val, X_test = features[:train_split], features[train_split:val_split], features[val_split:-1]
    Y_train, Y_val, Y_test = target[1:train_split+1], target[train_split+1:val_split+1], target[val_split+1:]

    #print shape of samples
    print('The shapes of the X features are:', X_train.shape, X_val.shape, X_test.shape)
    print('The shapes of the Y target are:', Y_train.shape, Y_val.shape, Y_test.shape)
    
    return X_train, X_val, X_test, Y_train, Y_val, Y_test
```

A negative R-squared value typically indicates that the model's fit is worse than a simple horizontal line (e.g., the mean or average of the dependent variable). This can happen when the model is not capturing the underlying patterns or relationships in the data and is performing poorly.

There are several possible reasons for a negative R-squared value:

Overfitting: If the model is overly complex or has too many variables compared to the amount of data available, it may overfit the training data and perform poorly on unseen data, resulting in a negative R-squared value.

Inappropriate model choice: Linear regression may not be appropriate for certain types of data or relationships. If the underlying data has nonlinear patterns or complex interactions, using a different modeling approach (e.g., polynomial regression, decision trees, or neural networks) may be more suitable.

Random Forest Regression: Random Forest is an ensemble learning method that combines multiple decision trees to make predictions. It can handle non-linear relationships, capture interactions between features, and handle missing values. Random Forest can be effective for stock price prediction by considering various features and their importance.
